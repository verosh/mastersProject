```python
#we want list of dicts with same headers
import json, os
import re

def jsonToDict(file):
    with open(file) as json_file:
        data = json.load(json_file)

        print(file, len(data['people']))
 
        d = data['people'][0]

        #print(type(d['hand_left_keypoints_2d']), type(d['hand_right_keypoints_2d']),type(d['pose_keypoints_2d']))
   
        f = { 'participant':  re.findall("(p_[0-9]+)",file)[0],'frame': int(re.findall("_([0-9]+)_", file)[0]) , 'pose_keypoints_2d' : d['pose_keypoints_2d'], 'hand_left_keypoints_2d': d['hand_left_keypoints_2d'], 'hand_right_keypoints_2d':d['hand_right_keypoints_2d']}
    return f

path_to_json = '/home/vero/coding/masters/mastersProject/src/output/p_109'
# import all files from folder which ends with .json 
json_files = glob.glob(os.path.join(path_to_json, '*.json'))
all_frames = []

for i in json_files:
    all_frames.append( jsonToDict(i) )

```

```python
print(len(all_frames))
```

```python
import csv 

field_names = [ "participant", "frame", "pose_keypoints_2d", "hand_left_keypoints_2d", "hand_right_keypoints_2d" ]


with open('poseHandsKeypoints.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(all_frames)
```
