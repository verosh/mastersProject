```python
import os
import pandas as pd
```

```python
folder = "/home/vero/Dropbox/wellbeingdataset"
subfolders = [ f.name for f in os.scandir(folder) if f.is_dir() ]
participants = [ int(i[2:]) for i in subfolders if i != 'Weizhe - features' and i != 'processed_derivations' ]
participants.sort()
print(participants)
```

```python
sheet = pd.read_excel("/home/vero/Dropbox/wellbeingdataset/interview_questionnaire_responses.xlsx")
```

```python
sheet
```

```python
#sheet['Participant ID'][participants]
df = sheet.loc[sheet['Participant ID'].isin(participants)].sort_values('Participant ID')
df = df.loc[ df['Notes']!='Failed to record video']
df
```

```python
print((df['Participant ID'].sort_values().to_list()))
print((participants))
```

```python
df.shape
```

```python
df = df.replace('Not at all', 0).replace('Several days', 1).replace('More than half the days', 2).replace('Nearly every day', 3)
```

```python
depression = df.loc[:, 'Little interest or pleasure in doing things':'Moving or speaking so slowly that other people could have noticed. Or the opposite - being so fidgety or restless that you have been moving around a lot more than usual']
depression['Participant ID'] = df['Participant ID']
depression
```

```python
anxiety = df.loc[:,'Feeling nervous, anxious or on edge':'Feeling afraid as if something awful might happen']
anxiety['Participant ID'] = df['Participant ID']
anxiety
```

```python
anxiety['sum'] = anxiety.loc[:,'Feeling nervous, anxious or on edge':'Feeling afraid as if something awful might happen'].sum(axis='columns')
anxiety
```

```python
depression['sum'] = depression.loc[:, 'Little interest or pleasure in doing things':'Moving or speaking so slowly that other people could have noticed. Or the opposite - being so fidgety or restless that you have been moving around a lot more than usual'].sum(axis='columns')
depression
```

```python
truths = pd.DataFrame()
truths['Participant ID'] = anxiety['Participant ID']
truths['GAD Score'] = anxiety['sum']
truths['PHQ Score'] = depression['sum']
truths
```

```python
truths.to_csv('truths.csv')
```
